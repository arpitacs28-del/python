```python
n = int(input("enter a number"))
number = int(n)
if number == 0:
    digit_count = 1
else:
    if number<0:
        number = -number
        digit_count = 0    
    while number>0:
     number = number//10
     digit_count = digit_count+1
    print("the number of digits is:",digit_count)
```

```python
balance = 10000
while True:
     print("1. Check balance")
     print("2. Deposit money")
     print("3. Withdraw money")
     print("4.Exit")

     choice = int(input("enter your choice:"))
     if choice==1:
          print("balance:",balance)
     elif choice==2:
          amount = int(input("enter the amount"))
          balance+=amount
          print("amount deposited:") 
     elif choice==3:
          amount = int(input("enter the amount:"))
          if amount<=balance:
             balance-=amount
             print("withdrawal successful")
          else:
               print("insufficient balance:")
     elif choice == 4:
          print("thank you")
          break
     else:
          print("invalid choice")          
```

```python
#prime numbers
num = int(input("enter a number"))
if num<2:
    print("not a prime number:")
else:
    for i in range(2,num):
        if num%i==0:
            print("not a prime number")
            break
    else:
        print("prime number")
```

```python
#reverse of a number
num = int(input("enter a number"))
orig = num
reverse = 0
while num>0:
    digit = num%10
    reverse = reverse*10+digit
    num = num//10
    print("original number:",orig)
    print("reversed number:",reverse)
```

```python
#palindrome number
num = int(input("enter a number"))
orig = num
reverse = 0
while num>0:
  digit = num%10
  reverse = reverse*10+digit
  num = num//10
  print("original number:",orig)
  print("reversed number:",reverse)
if orig==reverse:
        print("this is palindrome number")
else:
        print(" this is not a palindrome number")    
```

```python
#armstrong number
num = int(input("enter a number:"))
original = num
digits = len(str(num))
sum_of_powers = 0
while num>0:
    digit = num%10
    sum_of_powers += digit **digits
    num = num//10
    if original == sum_of_powers:
        print("Armstrong number:")
    else:
        print("not a armstrong number")    
```

```python
#fibonacci series
n = int(input("enter a number"))
a = 0
b = 1
for i in range(n):
    print(a,end=" ")
    c=a+b
    a=b
    b=c
    
```
